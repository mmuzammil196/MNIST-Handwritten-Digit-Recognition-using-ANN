# **Importing the Libraries**

```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

# **Download the Dataset from Keras Library**

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

# **Explore the Dataset**

```python
# shape of X_train dataset
X_train.shape
```

```python
# shape of X_test dataset
X_test.shape
```

```python
# shape of y_train dataset
y_train.shape
```

```python
# shape of y_test dataset
y_test.shape
```

# **Standarize the X_train & X_test Data in range between 0 & 1**

```python
X_train = X_train / 255
X_test = X_test / 255
```

# **Target Feature Check and Plot Digits Images**

# **Model Building - Train the Model**

```python
model = Sequential()

model.add(Flatten(input_shape=(28, 28)))
# input nodes
model.add(Dense(128, activation = 'relu'))
# output nodes and use softmax for multi classification problem
model.add(Dense(10, activation = 'softmax'))
```

```python
# model summary
model.summary()
```

```python
# compilation of model. use sparse categorical cross entropy for not apply one hot encoding when using this method
model.compile(loss='sparse_categorical_crossentropy', optimizer = 'Adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs = 10, validation_split = 0.2)
```

# **Visualize the Loss of Both Train and Test**

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

# **Visualize the Accuracy of Both Train and Test**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
```

# **Model - Prediction**

```python
y_prob = model.predict(X_test)
y_pred = y_prob.argmax(axis = 1)
```

```python
# model predict digit 7 of x_test data
y_pred[0]
```

# **Visualize the Prediction Digit Image**

```python
plt.imshow(X_test[0])
```

# **Model - Accuracy - Check**

```python
# accuracy of model is 97 %
accuracy_score(y_test, y_pred)
```
